```python
topic_sizes, topic_nums = top2vec.get_topic_sizes()
print(topic_sizes)
print("-----------------")
print(topic_nums)
```

```python
topic_words, word_scores, topic_nums = top2vec.get_topics(127)
print(topic_words)
```

```python
topic_words, word_scores, topic_scores, topic_nums = top2vec.search_topics(keywords=["infection"], num_topics=1)
print(topic_words)
print("----------")
print(word_scores)
print("----------")
print(topic_scores)
print("----------")
print(topic_nums)
print("----------")
```

```python
topic_words, word_scores, topic_scores, topic_nums = top2vec.search_topics(keywords=["sports"], num_topics=1)
for topic in topic_nums:
  top2vec.generate_topic_wordcloud(topic)
```

```python
documents, document_scores, document_ids = top2vec.search_documents_by_topic(topic_num=9, num_docs=5)
for doc, score, doc_id in zip(documents, document_scores, document_ids):
    print(f"Document: {doc_id}, Score: {score}")
    print("-----------")
    print(doc)
    print("-----------")
    print()
```

```python
documents, document_scores, document_ids = top2vec.search_documents_by_keywords(keywords=["Shinzo", "Abe"], num_docs=5)
for doc, score, doc_id in zip(documents, document_scores, document_ids):
    print(f"Document: {doc_id}, Score: {score}")
    print("-----------")
    print(doc)
    print("-----------")
    print()
```

```python
words, word_scores = top2vec.similar_words(keywords=["positive"], keywords_neg=[], num_words=40)
for word, score in zip(words, word_scores):
  print(f"{word} {score}")
```
